# Figure 2: Synthetic datasets

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../')
from dendrosplit import split,merge,utils
from dendrosplit_pipeline import *
%matplotlib inline
```

```python
# Analysis
def analyze_dataset(X,genes,dataset,D,sthresh=30,mthresh=10,verbose=True):
    start_analysis = time.time()
    ys,shistory,ym,mhistory = get_clusters(D,X,sthresh=sthresh,mthresh=mthresh,verbose=verbose,
                                           split_evaluator = split.select_genes_using_Welchs) # Get clusters
    print('analysis done (%.2f s elapsed)'%(time.time()-start_analysis))
    return X,genes,D,ys,shistory,ym,mhistory
```

```python
# Load data
dataset = 'Aggregation'
X = np.loadtxt('../data/synthetic_datasets/%s.txt'%(dataset))
X,Y = X[:,:2], X[:,2].astype(int)
np.random.seed(5)
X += np.random.normal(0,0.1,np.shape(X))
genes = np.array(range(2)).astype(str)

# Run DendroSplit
D = pairwise_distances(X)
D = (D+D.T)/2
X,genes,D,ys,shistory,ym,mhistory = analyze_dataset(X,genes,'',D,sthresh=40,
                                                    mthresh=10,verbose=False)
split.visualize_history(X,X[:,0],X[:,1],genes,shistory,markersize=10,axisoff=True)

# Plot final clusters
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(X[:,0],X[:,1],'.',c='k',label='',markersize=10)
plt.axis('off')
plt.title('%s dataset'%(dataset))
plt.subplot(1,2,2)
split.plot_labels_legend(X[:,0],X[:,1],ys,markersize=10)
_ = plt.title('DendroSplit clusters')
```

```python
from sklearn import datasets

# Generate noisy circles dataset
noisy_circles = datasets.make_circles(n_samples=500, factor=.5,
                                      noise=.05, random_state=0)
X,Y = noisy_circles[0],noisy_circles[1]
genes = np.array(range(2)).astype(str)

# Map points to distance from center
Xse = np.linalg.norm(X,axis=1).reshape(-1,1)

# Processing
D = pairwise_distances(Xse)
D = (D+D.T)/2
Xse,genes,D,ys,shistory,ym,mhistory = analyze_dataset(Xse,genes,'',D,
                                                      sthresh=30,mthresh=10,verbose=True)

plt.figure(figsize=(16,4))
plt.subplot(1,3,1)
split.plot_labels_legend(X[:,0],X[:,1],Y)
plt.title('True labels')
plt.subplot(1,3,2)
split.plot_labels_legend(Xse[:,0],Xse[:,0],Y)
plt.title('True labels after mapping')
plt.subplot(1,3,3)
split.plot_labels_legend(X[:,0],X[:,1],ys)
plt.title('DendroSplit labels')
```
